```python
import glob
from bs4 import BeautifulSoup as bs
import os
import pandas as pd
import timeit
```

```python
files = glob.iglob('./card/**/*.html')
```

```python
start = timeit.timeit()

big_list = []
i = 0
for f in files:
    small_list = []
    with open(f, 'rb') as file:
        soup = bs(file, 'html.parser')

    small_list.append(soup.title.get_text())
    small_list.append(soup.dd.get_text())
    
    big_list.append(small_list)
    i += 1

end = timeit.timeit()
print(end - start)
```

```python
df = pd.DataFrame(big_list, columns = ['Word', 'Definition'])
```
